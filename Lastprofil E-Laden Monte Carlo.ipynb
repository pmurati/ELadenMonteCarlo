# Lastprofil E-Laden Monte Carlo
$N=100$ Monte Carlo Simulationen über die Ladedauer und Ladeleistung von 25 PKWs über ein Jahr mit Drosselung ab variablem Schwellenwert: 80, 90, 100, 110, 120 kW.

## I: Annahmen der MC Simulation

Pro MC Iterarion werden folgende Parameter zufällig gesetzt.
- Jedem Nutzer wird eine fixe Energiemenge in kWh zugeordnet die er verbraucht. Diese wird zufällig aus einer Exponentialverteilung mit Parameter $\lambda=10$ gezogen.
- Basierend auf der Energiemenge wird für jeden Tag des Jahres eine zufällige Leistung in kW pro PKW initiiert. Es wird ein Grenzwert entsprechend dem 66% Quantil der Energiemengen aller Nutzer gesetzt. Die Nutzer im oberen Drittel laden mit einer fixen Leistung von 10kW, alle anderen mit einer täglich variierenden Leistung, gezogen aus einer Gleichverteilung im Interval $[3.5,10]$ kW. Simuliert werden damit zum Einen die Fahrtstrecken von Pendlern (Langstreckenfahrten) und zum anderen lokale, innerstädtische Arbeitswege.

```python
#Visualisierung der eben genannten Annahmen

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')

lam = 10
xx = np.arange(0,15,0.5)
yy = 1/lam * np.exp(-xx/lam)

p = 2/3
quantile = -np.log(1-p)*lam
qx = xx[xx>quantile].min()
qy = 1/lam * np.exp(-qx/lam)

plt.plot(xx,yy,linewidth=2,color='black')
plt.vlines(qx, ymin = 0, ymax = qy,
           colors = 'red', linestyle='--',
           label = '66% Quantil')

qxx1 = np.arange(0,qx+0.5,0.5)
qyy1 = 1/lam * np.exp(-qxx1/lam)
plt.fill_between(qxx1,qyy1,alpha=0.5,color='grey')
qxx2 = np.arange(qx,15,0.5)
qyy2 = 1/lam * np.exp(-qxx2/lam)
plt.fill_between(qxx2,qyy2,alpha=0.5,color='red')
plt.text(11.2,0.013,'P = 10 kW',fontsize=15)
plt.text(2,0.035,'P ~ U(3.5,10) kW',fontsize=15)
plt.title('Energie und Leistungsverteilung', fontsize=20)
plt.xlabel('Energie E in [kWh]',fontsize=15)
plt.ylabel('Wkt.Dichte ' + r'$p(E)$',fontsize=15)
plt.ylim([0,0.15])
plt.xlim([0,xx.max()])
plt.legend(fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.savefig(img_dir + 'energie_leistungsverteilung.png')
```

- Nutzer kommen unabhängig voneinander und gleichverteilt im Zeitinterval 16 - 20 Uhr von Ihrer Arbeit zurück und starten das Laden Ihrer PKWs. 
- Das Fahrverhalten aller Nutzer ändert sich an Wochenenden: 20% der Nutzer werden zufällig gezogen um Langstreckenausflüge zu tätigen. Energie und Leistung wird fix auf 100kWh und 10kW gesetzt, die Ankunftszeit liegt im uniformen Zeitfenster 16 - 20 Uhr. Die restlichen 80% erledigen Einkäufe, innerstädtische Fahrten etc. mit uniformer Energie, Leistung und Ankunftszeit in den entsprechenden Intervalen: $[0,17]$ kWh, $[3.5,10]$ kW und 9 - 20 Uhr. 
>**Anmerkung:** Eine Energiemenge von 0 kWh entspricht dem Fall, dass das Fahrzeug nicht genutzt wird.  
- Jeder Nutzer macht 30 Tage Urlaub im Jahr. Das Urlaubsstartdatum wird aus einer Normalverteilung gezogen, mit $\mu=211$ und $\sigma=\frac{\mu}{4}$. Dies entspricht dem Start der Sommerferien (211er Tag des Jahres ~ 30.07.2020 war der Beginn der Sommerferien in BW). Der Urlaub wird am Stück genommen. Urlaubsintervalle, die das Jahr überschreiten werden entsprechend auf Anfang und Ende des Jahres aufgeteilt. 

```python
import pandas as pd

day_range = pd.date_range(start='01.01.2020',end='31.12.2020',freq='1d')
xx = np.arange(0,len(day_range))
mu = 211
sigma = mu/4
yy = 1/(sigma*np.sqrt(2*np.pi)) *np.exp(-0.5*((xx-mu)/sigma)**2)

plt.plot(day_range,yy,color='black')

plt.title('Verteilung des Urlaubsstarts', fontsize=20)
plt.xlabel('Datum',fontsize=15)
plt.ylabel('Wkt.Dichte ' + r'$p(Datum)$',fontsize=15)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.savefig(img_dir + 'urlaubsstart_verteilung.png')
```

## II: Programstruktur

Es werden im Folgenden die einzelnen Codepassagen näher erläutert.

Zuerst werden die relevanten Pakete in die Session geladen.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
import os
import joblib

plt.style.use('seaborn-darkgrid')
results_dir = './results/'
img_dir = './images/'
```

Es wird die Funktion zur Bestimmung der energieabhängigen Leistung definiert (siehe **I: Annahmen der MC Simulation**)

```python
def power_calc(user_energy):
    user_power = np.zeros(len(user_energy))
    #Wähle Grenzwert entsprechend dem 66% Quantil
    thres = np.quantile(user_energy,2/3)
    
    for i in range(len(user_energy)):
        if user_energy[i] > thres:
            user_power[i] = 10
        else:
            user_power[i] = np.random.uniform(3.5,10)
            
    return(user_power)        
```

Initialisiere Parameter der Simulation:
- Anzahl der Iterationen `N=100`
- Anzahl der Nutzer `n_user=25`
- Gesamte Laufzeit über ein Jahr in 15min Auflösung `full_range`
- Zeitliche Auflösung `delta_t=0.25`
- Gesamte Laufzeit über ein Jahr in täglicher Auflösung `day_range`
- Array der einzelnen Drosselungsgrenzwerte `power_thres=80,90,100,110,120` 

```python
N = 100
n_user = 25
user = np.arange(n_user)

full_range = pd.date_range(start='01.01.2020 00:00:00',end='31.12.2020 23:5:00',freq='15min')
delta_t = 0.25

day_range = pd.date_range(start='01.01.2020',end='31.12.2020',freq='1d')

power_thres = np.arange(80,130,10)
```

Definiere numpy Arrays für die Ergebnisse der Simulation:
- Gesamtleistung bei keiner Drosselung `samples_unr`
- Dictionary mit einzelnen numpy Arrays pro Drosselungswert für die Gesamtleistung `samples_r`
- Dictionary mit einzelnen boolschen numpy Arrays pro Drosselungswert für das Ereignis einer Drosselung `above_thres`  

```python
samples_unr = np.zeros((len(full_range),N))

samples_r = {}
above_thres = {}
for p_thr in power_thres:
    samples_r[p_thr] = np.zeros((len(full_range),N))
    above_thres[p_thr] = np.zeros((len(full_range),N))
```

Start der Simulation:

Durchlaufe für jede Iteration $n=0,1,...,N$ die folgenden Schritte.

Setze die zu ladenden Energiemengen und die Urlaubszeiten.

Für jeden Tag des Jahres $day \in$ `day_range`:

Definiere eine neue lokale Variable mit den fixen Energien (Energiewerte können überschrieben werden, falls der tag auf's Wochenende fällt) und wähle das Zeitinterval aus `full_range`, welches dem jew. Tag entspricht. Setze die Ankunftszeit und berechne die Leistungsmenge aus den Energien für jeden Nutzer.

Prüfe, ob der Tag auf das Wochenende fällt und setzte ggf. Energie und Leistung neu. Bestimme die resultierende Ladezeit. 

Nun wird Leistungsmenge aller Nutzer für den jeweiligen Tag und mit der eben bestimmten Ladedauer zur Gesamtleistung aufsummiert und zum Array `samples_unr` für den jew. Tag und die jew. Iteration hinzugefügt. Prüfe für jeden Nutzer, ob er im Urlaub ist. Wenn nicht, so führe die eben genannte Summation durch.

Zuletzt, prüfe für die verschiedenen Drosselungswerte und für jeden Zeitschritt, ob der jew. Grenzwert überschritten wurde. Wenn ja, wird die überschüssige Leistung solange auf die nächsten zeitschritte umgeschichtet, bis es zu keiner Drosselung mehr kommt. 

>**Anmerkung:** Die Laufzeit des Skriptes beträgt mit den gewählten Parametern für `N` und `power_thres` ca. 1 Stunde. Das Skript kann übersprungen und die Ergebnisse im darauffolgenden Codeblock einfach in die Session reingeladen werden.

```python
#----------------
#aktuelles Skript
#----------------

np.random.seed(42) #zur Reproduzierbarkeit

for n in tqdm_notebook(range(N)):
    
    user_energy_fixed = np.random.exponential(10,n_user)
    
    #-------------
    #Urlaubszeiten
    #-------------
    #setze Urlaubszeiten fest, normalverteilt um die Sommerferien
    mu = 211 #30.07.2020 war Beginn der Sommerferien in BW 
    sigma = mu/4
    user_holiday = np.random.normal(mu,sigma,n_user).astype(int)

    for idx,holiday in enumerate(user_holiday):
        if holiday <0:
            user_holiday[idx] = 0
        if holiday>(len(day_range)-1):
            user_holiday[idx] = len(day_range) - 1   

    holiday_matrix = np.zeros((30,n_user))
    for u in range(n_user):
        holiday_matrix[:,u] = np.arange(user_holiday[u],user_holiday[u]+30)

    #falls das Jahr überschritten wird, wird der Urlaub am Anfang des Jahres genommen    
    holiday_matrix[holiday_matrix>(len(day_range)-1)] -= len(day_range)    
    holiday_matrix = holiday_matrix.astype(int) 
    
    #Durschlauf für jeden Tag des Jahres
    for d_idx,day in enumerate(day_range):
        
        #an Wochenenden wird andere Energiemenge verbraucht
        #setzte deshalb neue Variable, die am Anfang der Schleife mit fixen
        #Energien initialisiert wird
        user_energy = user_energy_fixed.copy()
        
        #filter nach den 96 15min Intervalen für den jew. Tag des Jahres
        restrict = full_range.date==day        
        restrict_index = np.where(restrict)[0]
        
        #ein Tag mit 15min Auflösung hat 96 Werte.
        #16:00 Uhr entspricht dem 64en Wert
        #20:00 Uhr entspricht dem 80en Wert
        #Werte aus diesem ganzzahligen Interval werden zufällig gezogen pro Fahrzeug        
        arrival = np.random.randint(64,80,n_user) 
        
        user_power = power_calc(user_energy)
        
        #----------
        #Wochenende
        #----------
        if day.dayofweek>=5:
            p_trip = np.random.random(n_user)
            
            #20% machen Langstreckenausflug
            long_trip = np.where(p_trip<=0.2)
            user_energy[long_trip] = 100
            user_power[long_trip] = 10
            
            short_trip = np.where(p_trip>0.2)
            user_energy[short_trip] = np.random.uniform(0,17,len(short_trip[0]))
            user_power[short_trip] = np.random.uniform(3.5,10,len(short_trip[0]))
            #die anderen 80% laden am WE im Zeitfenster 9Uhr (36er Wert) bis 20 Uhr
            arrival[short_trip] = np.random.randint(36,80,len(short_trip[0]))
        
        #bestimme gesamtzeit des Ladevorgangs
        user_time = (user_energy/(user_power*delta_t)).round(0)
        
        #------------------------------
        #Gesamtleistung ohne Drosselung
        #------------------------------
        for u in user:
            #es wird nur gefahren, wenn der Zeitpunkt nicht in die Urlaubszeit fällt
            if d_idx not in holiday_matrix[:,u]:
                #index_range = restrict_index[arrival[u]:(arrival[u]+int(user_time[u]))]
                #samples[index_range,n] += user_power[u]
                index_start = restrict_index[arrival[u]]
                index_end = restrict_index[arrival[u]]+int(user_time[u])
                samples_unr[index_start:index_end,n] +=user_power[u]
        
        
        #-----------------------------
        #Gesamtleistung mit Drosselung
        #-----------------------------
        #bestimme Gesamtleistung für alle Drosselungswerte
        for p_thr in power_thres:
            above_thres[p_thr][:,n] = samples_unr[:,n]>p_thr
            samples_r[p_thr][:,n] = samples_unr[:,n].copy()
            power_overload = 0
            #wenn Grenzwert überschritten wird, nimm die überschüssige Leistung und schichte
            #diese solange auf nachfolgende Zeitschritte um, bis es zu keiner Drosselung
            #mehr kommt
            for t in range(samples_unr.shape[0]-1):
                if samples_r[p_thr][t,n]>p_thr:
                    power_overload = samples_r[p_thr][t,n] - p_thr
                    samples_r[p_thr][t,n] = p_thr
                    samples_r[p_thr][t+1,n] += power_overload
```

Abspeichern/Reinladen der Ergebnisse.

```python
if 'MC_res.pkl' not in os.listdir(results_dir):
    MC_res = {}
    MC_res['samples_unrestricted'] = samples_unr
    MC_res['samples_restricted'] = samples_r
    MC_res['above_threshold'] = above_thres

    joblib.dump(MC_res,'MC_res.pkl')
else:
    MC_res = joblib.load(results_dir + 'MC_res.pkl')
    
    samples_unr = MC_res['samples_unrestricted']
    samples_r = MC_res['samples_restricted']
    above_thres = MC_res['above_threshold']
```

# Auswirkung auf Trafoebene

Betrachte die kummulierte Leistung von 8 Hauseinheiten (mit je 25 Nutzern) mit verschiedenen Drosselungen, angeschlossen an einer Trafostation mit jew. 600, 700 und 800 kW via bootstraping aus den MC Simulationen der gedrosselten Leistungsverläufe.

```python
N_trafo = 50
n_units = 8
power_thres_trafo = np.array([600,700,800])
samples_trafo = {}
above_thres_trafo = {}

for p_thr in power_thres:
    samples_trafo[p_thr] = np.zeros((len(full_range),N_trafo))
    for p_thr_t in power_thres_trafo:
        above_thres_trafo[p_thr,p_thr_t] = np.zeros((len(full_range),N_trafo))
```

```python
np.random.seed(42)

for n in tqdm_notebook(range(N_trafo)):
    for p_thr in power_thres:
        bs_index = np.arange(N)
        np.random.shuffle(bs_index)
        samples_trafo[p_thr][:,n] = samples_r[p_thr][0:len(full_range),bs_index[0:n_units]].sum(axis=1)
        for p_thr_t in power_thres_trafo:
            above_thres_trafo[p_thr,p_thr_t][:,n] = samples_trafo[p_thr][:,n] > p_thr_t
```

```python
%matplotlib inline

run_idx = 1
p_thr = 100
p_thr_t = 600

plt.figure(figsize=(18,8))
plt.plot(full_range,samples_trafo[p_thr][0:len(full_range),run_idx],
         linewidth=0.8,alpha=0.7)
plt.scatter(full_range[above_thres_trafo[p_thr,p_thr_t][0:len(full_range),run_idx]>0],
            p_thr_t*above_thres_trafo[p_thr,p_thr_t][0:len(full_range),run_idx][above_thres_trafo[p_thr,p_thr_t][0:len(full_range),run_idx]>0],
            marker='x',color='red')
plt.ylabel('Leistung [kW]',fontsize=15)
plt.title('Trafoleistung aus MC Simulation: {}\n Drosselung der Einheiten ab {} kW\n Drosselung des Trafos ab {} kW'.format(
    run_idx+1,p_thr,p_thr_t),fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig(img_dir + 'MC_trafo_run_2.png')
```

```python
#-----------------------
#wird nicht mehr benötgt
#-----------------------
def power_update(user_power, load, power_delta):
    power_res = user_power.copy()
    power_filter = power_res[load]
    idx = np.arange(len(power_filter))
    
    power_sum = sum(power_filter)
    
    power_sort = np.sort(power_filter)[::-1]
    idx_sort = np.argsort(power_filter)[::-1]
    
    #um auf ursprüngliche Reihenfolge zurückzukommen
    idx_mapping = np.stack((idx,idx_sort))
    
    reduction_success = False
    for i in np.arange(1,len(power_sort)-1):
        power_sort[0:i] = power_sort[i+1]
        if power_sum - sum(power_sort) >= power_delta:
            reduction_success = True
            break
    
    #wenn selbst die Summe über den kleinste Wert zu groß ist, dann
    #setzte gleichen Wert für alle an  
    if not reduction_success:
        equal_val =  (power_sum - power_delta)/len(power_sort)
        power_sort = np.full(len(power_sort),equal_val)
        
    #sortiere den Index des sortierten Arrays in der Matrix idx_mapping
    #greife auf die erste Reihe zu, welche die ursprüngliche Ordnung des
    #Arrays ausgibt. Effektiv wird damit die Sortierung rückgängig gemacht
    power_res[load] = power_sort[idx_mapping[:,idx_mapping[1,:].argsort()][0,:]] 
    
    return(power_res)
```
